```python
import cv2
import matplotlib.pyplot as plt

# Read the image in grayscale mode
img = cv2.imread('Pearson.jpg')

# Convert the image to grayscale

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Resize the image
img = cv2.resize(img, (650, 500))

# Apply binary thresholding
_, threshold = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)

contours,heirarchy=cv2.findContours(threshold, cv2.THRESH_BINARY, cv2.CHAIN_APPROX_SOLID )

for i in range(len(contours)):
    hull=cv2.convixHull(contours[i])
    cv2.drawContours(img,[hull],-1,(0,255,0),2)
    area = cv2.contourArea(contours[i])
    print(f'Area of contour {i+1}: {area}')

plt.imshow(img, cmap='gray')

cv2.waitKey(0)

cv2.destroyAllWindows()

```
